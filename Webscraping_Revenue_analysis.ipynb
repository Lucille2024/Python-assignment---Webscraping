```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_oil_and_gas_companies_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
                  
```

```python
table = soup.find('table')
country_titles = [th.text for th in table.find_all('th')]

country_titles
```

```python
country_table_titles = [title.text.strip() if hasattr(title, 'text') else title.strip() for title in country_titles] 

print(country_table_titles)
```

```python
#importing titles into a Dataframe

import pandas as pd

df = pd.DataFrame(columns = country_table_titles)
df
```

```python
#adding rows and columns to the Dataframe
column_data = table.find_all('tr')
```

```python
for row in column_data: 
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
#adding the table data to the dataframe to create the table
for row in column_data[1:]: 
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
import pandas as pd
```

```python
#remove the unwanted data in square brackets (the unwanted data in the square brackets in the columns are references (hyperlinks) to the source of the information on wikipedia


def remove_square_brackets(value):
    return str(value).split('[')[0].strip()
columns_to_clean = ['2017', '2018', '2019', '2020', '2021', '2022']

for column in columns_to_clean:
    df[column] = df[column].apply(remove_square_brackets)
```

SHOW GLOBAL REVENUE FOR 2017 TO 2022  ON A LINE CHART AND SHOW 2022 UNITED STATES REVENUE AS A PERCENTAGE OF GLOBAL REVENUE ON A PIECHART

```python
import pandas as pd
import matplotlib.pyplot as plt



# Exclude the "Global" row from the DataFrame
df_countries = df[df['Country'] != 'Global']

# Calculate the global revenue for each year
global_revenue = df_countries.groupby('Country')[df.columns[2:]].sum().sum()

# Creating a DataFrame with global revenue for the years 2017 to 2022
years = [2017, 2018, 2019, 2020, 2021, 2022]
global_revenue_data = pd.DataFrame({'Year': years, 'Global Revenue': global_revenue})

# Plotting the global revenue on a line chart
plt.figure(figsize=(10, 6))
plt.plot(global_revenue_data['Year'], global_revenue_data['Global Revenue'], marker='o', linestyle='-', color='b')
plt.title('Global Revenue Over the Years (2017-2022)')
plt.xlabel('Year')
plt.ylabel('Global Revenue (in billion USD)')
plt.grid(True)
plt.show()

# Calculate the United States revenue as a percentage of global revenue for the year 2022
us_revenue_percentage = (df_countries[df_countries['Country'] == 'United States']['2022'].sum() / global_revenue_data['Global Revenue'].iloc[-1]) * 100

# Creating a pie chart to show US revenue as a percentage of global revenue for the year 2022
labels = ['United States', 'Rest of the World']
sizes = [us_revenue_percentage, 100 - us_revenue_percentage]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('United States Revenue Percentage of Global Revenue (2022)')
plt.show()
```
